```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import Select
from selenium.webdriver.common.alert import Alert
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# ============================Section.1 進入遊戲大廳============================

driver = webdriver.Chrome()
driver.get("https://dev.iplaystar.net/")   # 進入 DEV前台網頁

game_lobby = WebDriverWait(driver, 5).until(
    EC.element_to_be_clickable((By.XPATH, "/html")))
game_lobby.click()   
print("進入DEV前台大廳!")

# ============================Section.2 登入遊戲帳號============================

login_form = WebDriverWait(driver, 5).until(
    EC.element_to_be_clickable((By.XPATH, '//*[@id="login-form"]')))

input_acc = WebDriverWait(driver, 5).until(
    EC.element_to_be_clickable((By.XPATH, '//*[@id="login-form"]/div[2]/input')))
input_acc.clear()    # 清除帳號欄位資訊
input_pass = WebDriverWait(driver, 5).until(
    EC.element_to_be_clickable((By.XPATH, '//*[@id="login-form"]/div[3]/input')))   
input_pass.clear()   # 清除密碼欄位資訊

actions = ActionChains(driver)
actions.move_to_element(input_acc)    
actions.perform()
input_acc.send_keys("ivan_li")   # xxxxx = 個人使用者帳號
time.sleep(1)
input_acc.send_keys(Keys.TAB)   # 切換至密碼輸入欄位
input_pass.send_keys("@Koe1050259@")  # xxxxx = 個人密碼
time.sleep(1)

rem_acc = WebDriverWait(driver, 5).until(
    EC.element_to_be_clickable((By.XPATH, '//*[@id="login-form"]/div[4]/div/div/label')))
rem_acc.click()    # 勾選在這部電腦上記住我的帳號
time.sleep(2)

canrem_acc = WebDriverWait(driver, 5).until(
    EC.element_to_be_clickable((By.XPATH, '//*[@id="login-form"]/div[4]/div/div/label')))
canrem_acc.click()    # 取消勾選在這部電腦上記住我的帳號
time.sleep(1)

check_box = WebDriverWait(driver, 5).until(
    EC.element_to_be_clickable((By.XPATH, '//*[@id="login-form"]/div[6]')))
confirm_btn = driver.find_element(By.XPATH, '//*[@id="login-form"]/div[6]/div[1]/input')
actions = ActionChains(driver)
actions.move_to_element(confirm_btn)    
actions.perform()
confirm_btn.click()  # 確認並登入
print("已正常登入!")
time.sleep(2)

# ============================Section.3 進入PS遊戲大廳============================

lobby_icon = WebDriverWait(driver, 10).until(
    EC.element_to_be_clickable((By.XPATH, '//*[@id="SLOT"]/div[1]/div')))
actions = ActionChains(driver)
actions.move_to_element(lobby_icon)   # 將滑鼠游標移至"遊戲大廳Play now"點擊處
actions.perform()
lobby_icon.click()  # 進入遊戲大廳(跳出遊戲宣傳畫面_火焰金字塔) 
time.sleep(5)

driver.switch_to.window(driver.window_handles[1])  # 切換至遊戲大廳瀏覽器分頁

pyr_open_1 = WebDriverWait(driver, 10).until(
    EC.element_to_be_clickable((By.CLASS_NAME,'modal-body')))
pyr_open_1.click()  # 進入火焰金字塔主頁
print("進入火焰金字塔遊戲主頁!")
time.sleep(10)

driver.switch_to.window(driver.window_handles[2])  # 切換至火焰金字塔瀏覽器分頁

driver.close()  # # 關閉火焰金字塔瀏覽器分頁
print("退出火焰金字塔主頁!")
time.sleep(3)

driver.switch_to.window(driver.window_handles[1])  # 切換至遊戲大廳瀏覽器分頁
icon_close = WebDriverWait(driver, 10).until(
    EC.element_to_be_clickable((By.CLASS_NAME, 'icon-close')))
icon_close.click()
print("關閉遊戲宣傳畫面並進入遊戲大廳!")
time.sleep(3) 

# ============================Section.4 介面功能切換============================

lobby_slot = WebDriverWait(driver, 5).until(
    EC.element_to_be_clickable((By.CLASS_NAME, 'icon-slot')))
lobby_slot.click()
print("進入Slot遊戲大廳!")
time.sleep(2) 

lobby_card = WebDriverWait(driver, 5).until(
    EC.element_to_be_clickable((By.XPATH, '/html/body/nav/ul/li[3]/div/a/i')))
lobby_card.click()
print("進入Card遊戲大廳!")
time.sleep(2) 

item_favo = WebDriverWait(driver, 5).until(
    EC.element_to_be_clickable((By.CLASS_NAME, 'icon-favorite')))
item_favo.click()
print("進入我的最愛遊戲功能選單!")
time.sleep(2) 

item_pro = WebDriverWait(driver, 5).until(
    EC.element_to_be_clickable((By.XPATH, '/html/body/nav/ul/li[5]/div/a/i')))
item_pro.click()
print("進入Profile功能畫面!")
time.sleep(2) 

func_home = WebDriverWait(driver, 5).until(
    EC.element_to_be_clickable((By.CLASS_NAME, 'icon-home')))
func_home.click()
icon_close = WebDriverWait(driver, 5).until(
    EC.element_to_be_clickable((By.CLASS_NAME, 'icon-close')))
icon_close.click()
print("進入遊戲大廳首頁!")
time.sleep(3) 


driver.close()
print("瀏覽器已關閉，DEV前台自動化測試結束!")


```
